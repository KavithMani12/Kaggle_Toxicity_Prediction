```python
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem import Descriptors
from rdkit.Chem import PandasTools
from rdkit.ML.Descriptors import MoleculeDescriptors
import matplotlib.pyplot as plt 
```

```python
train = pd.read_csv("train_II.csv")
test = pd.read_csv("test_II.csv")
```

```python
# Creating data frame with all descriptions columns name 
descriptors = MoleculeDescriptors.MolecularDescriptorCalculator([x[0] for x in Descriptors._descList])
desc_names = descriptors.GetDescriptorNames() # Retrieve all names of descriptors in array format
descriptors_for_smiles_train = pd.DataFrame(columns=desc_names)
descriptors_for_smiles_train
```

```python
train
```

```python
# Splitting Id into chemical and assay Id
chemical = []
assay_id = []
for id_s in train.Id:    
    split_with_semi_colen = id_s.split(";")
    chemical.append(split_with_semi_colen[0])
    assay_id.append(int(split_with_semi_colen[1]))
```

```python
train["Chemical"] = chemical
train["Assay_ID"] = assay_id
```

```python
# Droping the Id
train = train.drop("Id",axis=1)
```

```python
# Calculating Description for each chemical in train data
i=0
for chemical in train["Chemical"]:
    if(i%100==0):
        print("Findind description for",i)
    descriptors = MoleculeDescriptors.MolecularDescriptorCalculator([x[0] for x in Descriptors._descList])
    mols = Chem.MolFromSmiles(chemical)
    descripton = list(descriptors.CalcDescriptors(mols)) # Get descriptors and convert to array format
    descriptors_for_smiles_train.loc[i] = descripton
    i = i + 1
```

```python
# Adding ASsay id, Expected to features
descriptors_for_smiles_train["Assay_ID"] = assay_id
descriptors_for_smiles_train["Expected"] = train.Expected
```

```python
descriptors_for_smiles_train.to_csv("train_with_descriptors.csv",index=False)
```
